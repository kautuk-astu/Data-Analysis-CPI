```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.ticker as ticker
import matplotlib.dates as mdates
```

```python
df = pd.read_csv('All_India_Index_Upto_Feb24.csv')
```

### Preprocessing

```python
len(df)
```

```python
df.head()
```

```python
df.describe()
```

```python
numeric_cols = df.select_dtypes(include='number').columns.difference(['index','Year'])
df['CPI'] = df[numeric_cols].mean(axis=1)
```

```python
df['Year_Month'] = pd.to_datetime(
    df['Year'].astype(str).str.strip() + '-' + df['Month'].str.strip(),
    format='%Y-%B', errors='coerce'
)
df = df.dropna(subset=['Year_Month'])
df = df.sort_values('Year_Month')
```

```python
df_pivot = df.pivot(index='Year_Month', columns='Sector', values='CPI')
df_plot = df_pivot[['Rural','Urban']]
df_plot = df_plot.dropna()
```

```python
len(set(df_plot.index))
```

```python
df_plot.head()
```

### Line Plot Showing CPI Trend

```python
plt.figure(figsize=(14,6))
plt.plot(df_plot.index, df_plot['Rural'], marker='o', label='Rural CPI', color='tab:blue')
plt.plot(df_plot.index, df_plot['Urban'], marker='s', label='Urban CPI', color='tab:orange')

plt.title('CPI Trend Over Time (Rural vs Urban)', fontsize=20)
plt.xlabel('Year', fontsize=20)
plt.ylabel('CPI Value', fontsize=20)

ax = plt.gca()

ax.xaxis.set_major_locator(mdates.YearLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
ax.xaxis.set_minor_locator(mdates.MonthLocator())

ax.set_ylim(bottom=0, top=200)
ax.yaxis.set_major_locator(ticker.MultipleLocator(40))
ax.yaxis.set_minor_locator(ticker.AutoMinorLocator(3))

ax.tick_params(axis='x', which='major', labelsize=16)
ax.tick_params(axis='y', which='major', labelsize=16)
ax.tick_params(axis='x', which='minor', labelsize=12)
ax.tick_params(axis='y', which='minor', labelsize=12)

ax.grid(which='major', linestyle='--', alpha=0.6)

plt.xticks(rotation=45)
plt.legend(fontsize=16)
plt.tight_layout()
plt.savefig('cpi_trend.png', dpi=300)
plt.savefig('cpi_trend.pdf', dpi=300)
plt.show()
```

### Scatter Plot

```python
plt.figure(figsize=(10,6))
ax = plt.gca()

plt.scatter(df_plot['Rural'], df_plot['Urban'],
            c='tab:blue', edgecolor='black', s=80, label='Urban CPI')
plt.scatter(df_plot['Rural'], df_plot['Rural'],
            c='tab:orange', edgecolor='black', s=80, label='Rural CPI')

plt.title('Rural vs Urban CPI Comparison', fontsize=20, pad=20)
plt.xlabel('Rural CPI', fontsize=20, labelpad=10)
plt.ylabel('Urban CPI', fontsize=20, labelpad=10)

ax.set_xlim(0, 200)
ax.set_ylim(10, 200)

ax.xaxis.set_major_locator(ticker.MultipleLocator(40))
ax.xaxis.set_minor_locator(ticker.MultipleLocator(10))
ax.yaxis.set_major_locator(ticker.MultipleLocator(40))
ax.yaxis.set_minor_locator(ticker.MultipleLocator(10))

ax.tick_params(axis='x', which='major', labelsize=16, width=1.5, length=7)
ax.tick_params(axis='y', which='major', labelsize=16, width=1.5, length=7)
ax.tick_params(axis='x', which='minor', labelsize=12, width=1.0, length=4)
ax.tick_params(axis='y', which='minor', labelsize=12, width=1.0, length=4)

ax.grid(which='major', linestyle='--', alpha=0.6)
ax.grid(which='minor', linestyle=':', alpha=0.0)

ax.axhline(y=0, color='black', linewidth=1.5, zorder=5)
ax.axvline(x=0, color='black', linewidth=1.5, zorder=5)

ax.spines['top'].set_color('black')
ax.spines['right'].set_color('black')

plt.legend(fontsize=16)

plt.tight_layout()
plt.savefig('cpi_scatter.png', dpi=300)
plt.savefig('cpi_scatter.pdf', dpi=300)
plt.show()
```

### Box Plot – Distribution / Volatility

```python
plt.figure(figsize=(8,6))
plt.boxplot([df_plot['Rural'].values, df_plot['Urban'].values],  # use .values to be safe
            labels=['Rural', 'Urban'], patch_artist=True,
            boxprops=dict(facecolor='skyblue', color='blue'),
            medianprops=dict(color='red'),
            whiskerprops=dict(color='blue'),
            capprops=dict(color='blue'),
            flierprops=dict(markerfacecolor='orange', marker='o', markersize=6, linestyle='none')
           )
plt.title('CPI Distribution & Volatility (Rural vs Urban)', fontsize=16)
plt.ylabel('CPI Value', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

plt.figure(figsize=(8,6))

# === Boxplot ===
plt.boxplot([df_plot['Rural'].values, df_plot['Urban'].values],
            labels=['Rural', 'Urban'],
            patch_artist=True,
            boxprops=dict(facecolor='skyblue', color='blue'),
            medianprops=dict(color='red'),
            whiskerprops=dict(color='blue'),
            capprops=dict(color='blue'),
            flierprops=dict(markerfacecolor='orange', marker='o', markersize=6, linestyle='none')
           )

# === Titles & Labels ===
plt.title('CPI Distribution & Volatility (Rural vs Urban)', fontsize=20, pad=15)
plt.ylabel('CPI Value', fontsize=20, labelpad=10)

# === Axes styling ===
ax = plt.gca()

# === Y-axis ticks ===
ax.yaxis.set_major_locator(ticker.MultipleLocator(40))
ax.yaxis.set_minor_locator(ticker.MultipleLocator(10))

# === X-axis ticks ===
ax.xaxis.set_major_locator(ticker.FixedLocator([1,2]))  # positions of boxes
ax.xaxis.set_minor_locator(ticker.NullLocator())         # no minor ticks on x-axis

# === Tick parameters ===
ax.tick_params(axis='y', which='major', labelsize=16, width=1.5, length=7)
ax.tick_params(axis='y', which='minor', labelsize=12, width=1.0, length=4)
ax.tick_params(axis='x', which='major', labelsize=16, width=1.5, length=7)

# === Gridlines ===
ax.grid(which='major', linestyle='--', alpha=0.6)
ax.grid(which='minor', linestyle=':', alpha=0.0)

plt.tight_layout()
plt.savefig('cpi_boxplot.png', dpi=300)
plt.savefig('cpi_boxplot.pdf', dpi=300)
plt.show()
```
